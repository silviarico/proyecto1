```python
def muertos(a):
    if str(a).lower()=="unknown":
        return "desconocido"
    if "y" in str(a).lower():
        return True
    elif "n" in str(a).lower():
        return False
    else:
        return "desconocido"
```

```python
def estaciones(a):
    if "07-10" in str(a):
        return "summer"
    elif "unknown" in str(a):
        return a
    elif int(a)<4 and int(a)>=1:
        return "winter"
    elif int(a)>=4 and int(a)<7:
        return "spring"
    elif int(a)>=7 and int(a)<10:
        return "summer"
    elif int(a)>=10 and int(a)<=12:
        return "autumn"
    else:
        return "unknown"
```

```python
at["seasons"]=at["mes"].apply(estaciones) #creo una nueva columna que agrupa las fechas por estaciones
```

```python
def edad(a):
    if "month" in str(a).lower():
        return 1
    elif "teen" in str(a).lower():
        return 15
    elif "middle" in str(a).lower() or "adul" in str(a).lower():
        return 40
    elif "elderly" in str(a).lower():
        return 65
    elif "young" in str(a).lower():
        return 10
    else:
        return a
```

```python
at["edad"] = at["age"].apply(edad) 
```

```python
def shark(a):
    if "white" in str(a).lower():
        return "white shark"
    elif "tip" in str(a).lower():
        return "coloured tip shark"
    elif "uncon" in str(a).lower() or "not conf" in str(a).lower() or "questionable" in str(a).lower():
        return "no evidence of shark involvement"
    elif "galap" in str(a).lower():
        return "galapagos shark"
    elif "nurse" in str(a).lower():
        return "nurse shark"
    elif "tiger" in str(a).lower():
        return "tiger shark"
    elif "bull" in str(a).lower():
        return "bull shark"
    elif "lemon" in str(a).lower():
        return "lemon shark"
    elif "bronze w" in str(a).lower():
        return "bronze whaler shark"
    elif "mako" in str(a).lower():
        return "mako shark"
    elif "hammer"in str(a).lower():
        return "hammerhead shark"
    elif "blue"in str(a).lower():
        return "blue shark"
    else:
        return "other"
```

```python
at["especies"] = at["species"].apply(shark)#agrupo los tiburones por raza
```

```python
count5=at["especies"].value_counts()
dict_shark1=count5.to_dict()
print(dict_shark1)
```

```python
def franja(a):
    if pd.isna(a)==True:
        return a
    if int(a)>=1 and int(a)<12:
        return "childs"
    elif int(a)>=12 and int(a)<19:
        return "teens"
    elif int(a)>=19 and int(a)<35:
        return "young_adult"
    elif int(a)>=35 and int(a)<55:
        return "adult"
    elif int(a)>=55 and int(a)<75:
        return "mature_adult"
    else:
        return "advanced_age"
    
```

```python
at["franja_edad"]=at["edad"].apply(franja) #creo franjas de edad

```
